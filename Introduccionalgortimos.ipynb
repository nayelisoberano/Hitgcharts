# **Características de los algoritmos**



1.   Entradas.(0 o mas entradas.)
2.   Salidas. (Al menos 0 o mas salidas.)
3.   Ser definido.(Todos los pasos tienen que ser claro y que puedan solucionarse)

4.   Finito.(El algoritmo debe de teminar en cualquier punto)
5.   Efectivo.

Todo algoritmo tiene que terminar en algún punto.

**Ejemplo:**

En un algoritmo no son importantes los detalles técnicos, propios de la implementación (tipo de variable , declaración de variable , el uso de arreglos , etc.)






```
#               Algoritmo ( a , b )
                      Inicio
                      tem <- a
                      a <--- b
                      b <- temp
                      end.
```




 La implementación de este algoritmo depende del programador que desarrolle un programa que implemente este algoritmo.

                   

````python
def intercambiar(a, b):
    temp = a
    a = b
    b = temp
    return a, b

# Ejemplo de uso
x = 5
y = 10
x, y = intercambiar(x, y)
print("Después del intercambio:")
print("x =", x)
print("y =", y)
````
